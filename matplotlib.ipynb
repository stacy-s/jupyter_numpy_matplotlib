## Основные функции для построения графиков:
* `plt.plot(x, y, *args)` — рисует график по точкам с координатами из $x$ по горизонтальной оси и из $y$ по вертикальной оси. Точки будут соединяться в том порядке, в котором они указаны в этих массивах. Построим график по заданным точкам $x$ и $y$:

```python
plt.plot(np.arange(0, 6, 1), np.arange(0, 6, 1))
```

* `plt.xlim(left, right)` — установить пределы графика по горизонтальной оси. `left` - левый предел, `right` - правый предел по оси x. Ось x будет отображаться от значения `left` до значения `right`.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.xlim(left=-np.pi, right=np.pi)
plt.plot(x, np.cos(x))
```

Без использования функции `plt.xlim(left, right)` график будет выглядить следующим образом:

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
```

Функцию можно использовать, указывая один параметр:

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.xlim(left=-np.pi)
plt.plot(x, np.cos(x))
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.xlim(right=-np.pi)
plt.plot(x, np.cos(x))
```

Так же функцию `plt.xlim()` можно использовать без указания параметров для получения текущих пределов графика по оси x.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.xlim()
```

* `plt.ylim(bottom, top)` устанавливает пределы графика по вертикальной оси. Единственное отличие от функции `plt.ylim` состоит в том что действия выполняются относительно вертикальной оси.

```python
x = np.linspace(-2, 2, 100)
plt.ylim(bottom=-1, top=1)
plt.plot(x, x**3)
```

Без использования функции график будет выглядить следующим образом:

```python
x = np.linspace(-2, 2, 100)
plt.plot(x, x**3)
```

Использование функции с одним параметром:

```python
x = np.linspace(-2, 2, 100)
plt.ylim(bottom=-0.1)
plt.plot(x, x**3)
```

Использование функции `plt.xlim(left, right)` и `plt.ylim(bottom, top)` для одного графика:

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.xlim(left=-np.pi, right=np.pi)
plt.ylim(bottom=-1.5, top=1.5)
```

`plt.title` устанавливает имя графика. Параметр `label` принимает текст названия графика. Так же у функции есть дополнительные параметры, которые можно посмотреть в официальной документации на сайте https://matplotlib.org/. 

```python
plt.plot(np.arange(0, 6, 1), np.arange(0, 6, 1))
plt.title('Линейная зависимость')
```

* `plt.xlabel(xlabel)` устанавливает название горизонтальной оси. Параметр `xlabel` принимает название оси x. Описание дополнительных параметром также представлено в документации.

```python
plt.xlabel(xlabel='x')
plt.plot(np.arange(0, 6, 1), np.arange(0, 6, 1))
```

* `plt.ylabel(ylabel)` устанавливает название вертикальной оси. Параметр `ylabel` принимает название оси y. Описание дополнительных параметров также представлено в документации.

```python
plt.ylabel(ylabel='f(x)')
plt.plot(np.arange(0, 6, 1), np.arange(0, 6, 1))
```

На одном графике можно отображать несколько функций. Отобразим функции $sin(x)$ и $cos(x)$:

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
```

В данном случае очевидно каким цветом нарисована функция $sin(x)$ и $cos(x)$, но чаще всего это не очевидно. С помощью функции `plt.legend()` можно сделать легенду. Но для того, чтобы `matplotlib` мог построить корректно легенду, необходимо в функции построения графика (в данном случае функция `plot`) передать текст легенды для данного графика в параметр `label`.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.sin(x), label='sin(x)')
plt.legend()
```

Иначе будет выведено предупреждение:

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
plt.legend()
```

Добавить сетку на график можно с помощью функции `plt.grid()`. Описание дополнительных параметров функции представлено в документации.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.grid()
plt.plot(x, np.cos(x))
```

Функция `plt.savefig(fname)` сохраняет график в файл с форматом `png` с именем, указанным в параметре `fname`.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.savefig(fname='cos')
```

`plt.show()` показывает график. Не рекомендуется использовать в интерактивном режиме (например в среде `jupyter`). В неинтерактивном режиме отобразит график в виде изображения (в неинтерактивном режиме без этой функции не будет отображаться график). 

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x, np.cos(x))
plt.show()
```

## Размещение графиков на разных полях
Все объекты в `matplotlib` имеют иерархическую (древоподобную) структуру. Основными элементами являются рисунок, панели и оси координат.


На верхнем уровне иерархии находится окно рисунка, которое является контейнером для остальных элементов. Рисунок является объектом класса __Figure__.
На рисунке может располагаться несколько (одна или больше) панелей - независимых областей рисования. Панели являются объектами класса __Axes__. Каждая из панелей содержит основные элементы рисунка: оси координат, метки на осях, текстовые метки, графики и т. д.

До этого момента мы не работали с __Figure__ напрямую и  значения ее параметров задавались по умолчанию. Для решения текущей задачи придется один из параметров – размер рисунка, задать вручную. За это отвечает аргумент `figsize` функции `plt.figure()`, которому присваивается кортеж из двух `float` элементов, определяющих высоту и ширину подложки.

После задания размера, указывается местоположение, куда будет установлено поле с графиком с помощью функции `plt.subplot()`. Чаще всего используют следующие варианты вызова функции `plt.subplot`:
 * `subplot(nrows, ncols, index)`, где `nrows` - обозначает количество строк, `ncols` - количество столбцов, `index` обозначает местоположение элемента. Зададим размер рисунка 3 х 6 и постороим два графика рядом:

```python
plt.figure(figsize=(6, 3))
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.subplot(1, 2, 1)
plt.plot(x, np.cos(x))
plt.subplot(1, 2, 2)
plt.plot(x, x)
```

 * `plt.subplot(pos)`, где pos - позиция, в виде трехзначного числа, содержащего информацию о количестве строк, столбцов и индексе, например 212, означает подготовить разметку с двумя строками и одним столбцов, элемент вывести в первую позицию второй строки. Этот вариант можно использовать, если количество строк и столбцов сетки не более 10, в ином случае следует обратиться к первому варианту.

```python
plt.figure(figsize=(3, 6))
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.subplot(211)
plt.plot(x, np.cos(x), color='#FF0000')
plt.subplot(212)
plt.plot(x, x, color='#0000FF')
```

```python
plt.figure(figsize=(12, 12))
x = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.subplot(221)
plt.plot(x, np.cos(x), color='#FF0000')
plt.subplot(222)
plt.plot(x, x, color='#0000FF')
plt.subplot(223)
plt.plot(x, x + 2, color='#00FF00')
plt.subplot(224)
plt.plot(x, x*x, color='#00FFFF')
```

 * Одно из неудобств использования последовательного вызова функций `plt.subplot` заключается в том, что каждый раз приходится указывать количество строк и столбцов сетки. Для того, чтобы этого избежать, можно воспользоваться функцией `plt.subplots`, в первых двух параметрах передается количество строк и столбцов сетки, В параметр `figure` передается размер всего рисунка. Функция `plt.subplots` возвращает два объекта, первый – это __Figure__, подложка, на которой будут размещены поля с графиками, второй – объект или массив объектов __Axes__, через которые можно получить полных доступ к настройке внешнего вида отображаемых элементов. 

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
x = np.linspace(-2*np.pi, 2*np.pi, 100)
axs[0, 0].plot(x, np.cos(x), color='#FF0000')
axs[0, 1].plot(x, x, color='#0000FF')
axs[1, 0].plot(x, x + 2, color='#00FF00')
axs[1, 1].plot(x, x*x, color='#00FFFF')
```

При работе с объектам класса `Axes` необходимо применять метод `Axes.set_xlim` вместо `plt.xlim`, а вместо `plt.ylim` соотвественно `Axes.set_ylim` параметры методов остаются прежними. Полный список методов класса `Axes` доступен в официальной документации https://matplotlib.org/3.1.1/api/axes_api.html.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
x = np.linspace(-2*np.pi, 2*np.pi, 100)
for i in range(2):
    for j in range(2):
        axs[i, j].set_ylim(bottom=-1, top=2)
axs[0, 0].plot(x, np.cos(x), color='#FF0000')
axs[0, 1].plot(x, x, color='#0000FF')
axs[1, 0].plot(x, x + 2, color='#00FF00')
axs[1, 1].plot(x, x*x, color='#00FFFF')
```

## Другие функции для построения графиков

Вместо функции `plt.plot` - можно использовать другие функции. 

* `plt.scatter(x, y, *args)` — рисует точки с координатами из $x$ по горизонтальной оси и из $y$ по вертикальной оси

```python
x = np.arange(1, 8, 1)
plt.scatter(x=x, y=1/x)
```

Большинство функций и параметров которые были расмотрены для рисования графика по точкам доступны и для функции `plt.scatter`

```python
x = np.arange(1, 8, 1)
plt.scatter(x=x, y=1/x, color='c', label='f(x)=1/x')
plt.scatter(x=x, y=2/x, color='r', label='f(x)=2/x')
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.xlim(left=0)
```

В пакете `numpy` есть функция генерации чисел из  распределения Пуассона - `numpy.random.poisson(lam=1.0, size=None)`, где `size` - размер выходного массива, если $size=None$, то будет возвращено одно число из нормального распределения. 

С помощью распределения Пуассона сгенерируем количество заболевщих людей в зависимости от температуры в городах __A__ и __B__.

```python
t_min, t_max = 0, 10
a = np.random.poisson(size=t_max+1)
b = np.random.poisson(size=t_max+1)
```

Предствим это дискретное распределение в виде точек:

```python
t = np.arange(t_min, t_max+1, 1)
plt.scatter(x=t, y=a, color='b', label='Город А')
plt.scatter(x=t, y=b, color='r', label='Город В')
plt.title('Зависимость заболевших от температуры')
plt.xlabel('температура воздуха')
plt.ylabel('Количество заболевших')
plt.legend()
```

Чтобы отследить зависмость лучше, воспользуемся непрерывным представление:

```python
t = np.arange(t_min, t_max+1, 1)
plt.plot(t, a, color='b', label='Город А')
plt.plot(t, b, color='r', label='Город В')
plt.title('Зависимость заболевших от температуры')
plt.xlabel('температура воздуха')
plt.ylabel('Количество заболевших')
plt.legend()
```

С помощью функции `plt.bar` - можно отобразить данный в виде столбчатой диаграммы. Отобразим полученные данные с помошью столбчатой диаграммы:

```python
t = np.arange(t_min, t_max+1, 1)
plt.bar(t, a, color='b', alpha=0.6, label='Город А')
plt.bar(t, b, color='r', alpha=0.6, label='Город В')
plt.title('Зависимость заболевших от температуры')
plt.xlabel('температура воздуха')
plt.ylabel('Количество заболевших')
plt.legend()
```

Параметр `alpha` в данном случае отвечает за прозрачность диаграммы и принимает значение в диапозоне $[0, 1]$. Также этот параметр часто применяется в функции `plt.scatter`, чтобы избежать эффекта слипания точек при большом их количестве на графике. 

Без изменения прозрачности точек:

```python
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
plt.scatter(x, y)
```

График с увеличенной прозрачность точек:

```python
plt.scatter(x, y, alpha=0.3)
```

## Построение графиков функции, неявляющихся непрерывными

При использование функции `plt.plot()` для построения функций с разрывами получается некорректный результат. Например при построение гиперболы y=1/x: 

```python
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(6, 6))
ax.grid()
ax.set_ylim(-10, 10)
ax.set_xlim(-10, 10)
ax.plot(x, 1 / x)
```

Так как `matplotlib` игнорирует (и не соединяет точки) числа типа `+inf`, `-inf`, `NaN`. нужно добиться, чтобы значение близкие к точке разрыва приняли значение `NaN`. Этого можно добиться с помощью функционала numpy массивов: 

```python
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(6, 6))
ax.grid()
ax.set_ylim(-10, 10)
ax.set_xlim(-10, 10)
x[np.argmin(np.abs(x))] = np.nan
ax.plot(x, 1 / x)
```
